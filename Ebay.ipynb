```python
ebay_df = pd.read_csv('EbayPcLaptopDataUnclean.csv')
ebay_price = pd.read_csv('EbayPcLaptopPriceData.csv')

pr_name_col = ebay_price.pop('Price')
ebay_df.insert(1, 'Price', pr_name_col)

# ebay_df = pd.merge(ebay_df, ebay_price, on="Item Number", how='left')
ebay_df
```

Based on the statistical results on the misisng vlaues, it can be found that "Rating", "Ratings Count", "Manufacturer Color", "Release Year", and "Country Region Of Manufacture" are the most missing part of this data frame. If we drop these, there is no significant impact on the 

```python
# Get the "Mode" from each column in case it can be used as the fill up the missing values.
def get_modes_for_columns(df, column_list):
    modes = {}
    for column in column_list:
        mode_value = df[column].mode().iloc[0] if not df[column].mode().empty else None
        modes[column] = mode_value
    return pd.Series(modes)

df = ebay_df
column_list = ['Brand',	
               'Price', 
               'Rating', 
               'Ratings Count', 
               'Processor', 
               'Screen Size', 
               'Manufacturer Color', 
               'Color', 'Ram Size', 
               'SSD Capacity', 
               'GPU', 
               'Processor Speed', 
               'Type', 
               'Release Year', 
               'Maximum Resolution', 
               'Model', 
               'OS', 
               'Hard Drive Capacity', 
               'Country Region Of Manufacture', 
               'Storage Type', 
               'Features', 
               'Condition', 
               'Seller Note']  

modes_series = get_modes_for_columns(df, column_list)
print("Modes for selected columns:")
print(modes_series)

```

```python
# Brand, Processor, OS, Model, Type should be shown since these are the most important part of a certain laptop

# Brand cannot be filled with based on known data, use unknown instead.
ebay_df['Brand'].fillna(value = 'Unknown', inplace=True)
# We cannot decide or put down a certian Processor data, since the series and base speed various a lot
ebay_df['Processor'].fillna(value = 'Unknown', inplace=True)
# Until know, the most popular OS in the list is Windows 11 Pro
ebay_df['OS'].fillna(value = 'Windows 11 Pro', inplace=True)
# We cannot make any predicition or guess the excat Model for different Makes
ebay_df['Model'].fillna(value = 'Unknown', inplace=True)
# Based on the title of the csv file, all the selling items are laptops
ebay_df['Type'].fillna(value = 'Laptop', inplace=True)
# Until know, we can still assume that the most popular size is 8GB
ebay_df['Ram Size'].fillna(value = '8GB', inplace=True)
ebay_df['Color'].fillna(value = 'Black', inplace=True)
ebay_df['Condition'].fillna(value = 'Unknown', inplace=True)
ebay_df['Maximum Resolution'].fillna(value='1920 x 1080', inplace=True)
```
